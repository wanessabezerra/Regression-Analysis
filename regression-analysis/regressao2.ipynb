```python
import pandas as pd
import statsmodels.api as sm
```

```python
dados = pd.read_csv('tabela_final.csv', sep=';')
```

```python
dados.columns
```

```python
# Somar os valores das colunas
soma_persistiu = dados['ALGORITMOS E LÓGICA DE PROGRAMAÇÃO'] + \
                 dados['INTRODUÇÃO À INFORMÁTICA'] + \
                 dados['FUNDAMENTOS DE MATEMÁTICA'] + \
                 dados['LÓGICA'] + \
                 dados['TEORIA GERAL DA ADMINISTRAÇÃO'] + \
                 dados['PROGRAMAÇÃO'] + \
                 dados['CÁLCULO DIFERENCIAL E INTEGRAL'] + \
                 dados['TEORIA GERAL DOS SISTEMAS'] + \
                 dados['PROGRAMAÇÃO ORIENTADA A OBJETOS I'] + \
                 dados['ESTRUTURA DE DADOS'] + \
                 dados['ÁLGEBRA LINEAR'] + \
                 dados['ORGANIZAÇÃO, SISTEMAS E MÉTODOS'] + \
                 dados['FUNDAMENTOS DE SISTEMAS DE INFORMAÇÃO'] + \
                 dados['PROGRAMAÇÃO WEB'] + \
                 dados['ARQUITETURA DE COMPUTADORES'] + \
                 dados['PROBABILIDADE E ESTATÍSTICA'] + \
                 dados['BANCO DE DADOS'] + \
                 dados['ENGENHARIA DE SOFTWARE I'] + \
                 dados['PROGRAMAÇÃO ORIENTADA A OBJETOS II'] + \
                 dados['SISTEMAS OPERACIONAIS'] + \
                 dados['PROJETO E ADMINISTRAÇÃO DE BANCO DE DADOS'] + \
                 dados['ENGENHARIA DE SOFTWARE II'] + \
                 dados['REDES DE COMPUTADORES'] + \
                 dados['CONTABILIDADE E CUSTOS'] + \
                 dados['EMPREENDEDORISMO EM INFORMÁTICA'] + \
                 dados['GESTÃO DE PROJETO DE SOFTWARE'] + \
                 dados['PROGRAMAÇÃO VISUAL'] + \
                 dados['MATEMÁTICA FINANCEIRA'] + \
                 dados['SISTEMAS DE APOIO À DECISÃO'] + \
                 dados['ÉTICA']

# Exibir a soma dos valores das colunas
print(soma_persistiu)
```

```python
dados['soma_persistiu'] = soma_persistiu
```

```python
dados['status'] = dados['status'].replace({'CANCELADO': 1, 'ATIVO': 0, 'CONCLUÍDO': 0})
dados
```

```python
y = dados['status']
chance_evasao = dados['soma_persistiu'] * 0.1 
y = chance_evasao
y
```

```python
dados['y'] = y
```

```python
dados = dados.drop(columns=['discente', 'ano_ingresso', 'ultimo_periodo'])
```

```python
#1. Preparação dos Dados
# Suponha que você tenha um dataframe chamado 'dados' com as colunas 'X1', 'X2', ..., 'Xn' para as reprovações em cada disciplina
# e a coluna 'Y' para a quantidade de vezes que o aluno persistiu no curso
# Exemplo:
# dados = pd.DataFrame({'X1': [2, 3, 1, 4], 'X2': [1, 2, 0, 3], 'Y': [5, 6, 4, 7]})
```

```python
# 2. Ajuste do Modelo de Regressão Linear
X = dados[['ALGORITMOS E LÓGICA DE PROGRAMAÇÃO', 'INTRODUÇÃO À INFORMÁTICA',
       'FUNDAMENTOS DE MATEMÁTICA', 'LÓGICA', 'TEORIA GERAL DA ADMINISTRAÇÃO',
       'PROGRAMAÇÃO', 'CÁLCULO DIFERENCIAL E INTEGRAL',
       'TEORIA GERAL DOS SISTEMAS', 'PROGRAMAÇÃO ORIENTADA A OBJETOS I',
       'ESTRUTURA DE DADOS', 'ÁLGEBRA LINEAR',
       'ORGANIZAÇÃO, SISTEMAS E MÉTODOS',
       'FUNDAMENTOS DE SISTEMAS DE INFORMAÇÃO', 'PROGRAMAÇÃO WEB',
       'ARQUITETURA DE COMPUTADORES', 'PROBABILIDADE E ESTATÍSTICA',
       'BANCO DE DADOS', 'ENGENHARIA DE SOFTWARE I',
       'PROGRAMAÇÃO ORIENTADA A OBJETOS II', 'SISTEMAS OPERACIONAIS',
       'PROJETO E ADMINISTRAÇÃO DE BANCO DE DADOS',
       'ENGENHARIA DE SOFTWARE II', 'REDES DE COMPUTADORES',
       'CONTABILIDADE E CUSTOS', 'EMPREENDEDORISMO EM INFORMÁTICA',
       'GESTÃO DE PROJETO DE SOFTWARE', 'PROGRAMAÇÃO VISUAL',
       'MATEMÁTICA FINANCEIRA', 'SISTEMAS DE APOIO À DECISÃO', 'ÉTICA']]  # Variáveis independentes (reprovações em cada disciplina)
X = sm.add_constant(X)  # Adiciona uma coluna de intercepto
y = dados['y']  # Variável dependente (quantidade de vezes que o aluno persistiu no curso)

# Ajusta o modelo de regressão linear
modelo = sm.OLS(y, X).fit()

# 3. Interpretação dos Coeficientes
print(modelo.summary())

# 4. Avaliação do Modelo
print("R²:", modelo.rsquared)
print("MSE:", modelo.mse_resid)
print("RMSE:", modelo.mse_resid ** 0.5)
# Calcular as previsões do modelo
previsoes = modelo.predict(X)

# Calcular o erro absoluto entre as previsões e os valores reais
erro_absoluto = abs(previsoes - y)

# Calcular a média do erro absoluto para obter o MAE
MAE = erro_absoluto.mean()

print("MAE:", MAE)

# 5. Previsão da Evasão (se necessário)
# Use o modelo para fazer previsões sobre a evasão de novos alunos
# por exemplo, usando modelo.predict(novos_dados)
```
